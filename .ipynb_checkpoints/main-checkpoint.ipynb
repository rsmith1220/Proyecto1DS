```python
# Find columns with null values
null_columns = df.isnull().sum()

# Display the result
print(null_columns)
```

Al usar los métodos isnull y sum del dataframe, podemos contar los respectivos valores nulos de cada columna del Dataframe. Las columnas con valores nulos son: Distrito, Dirección, teléfono, supervisor y director. Tales columnas son las que requieren tratamientos por nuestra parte. 

```python
df.drop('CODIGO', axis=1, inplace=True)
df.drop('TELEFONO', axis=1, inplace=True)
df.drop('DIRECTOR', axis=1, inplace=True)
df.drop('SUPERVISOR', axis=1, inplace=True)
```

Usamos el método drop para eliminar las columnas Código, teléfono, director y supervisor; al tener el parámetro inplace= True, modificamos el dataframe original eliminando del mismo dichas columnas.

```python
# Eliminar filas duplicadas en todo el DataFrame
df.drop_duplicates(inplace=True)
```

El método drop_duplicates nos permite eliminar columnas que contengan los mismos valores, i.e., que sean columnas duplicadas.

```python
# Fill null values in the 'column_name' column with the previous value in the column
df['DIRECCION'].fillna(method='ffill', inplace=True)
# Fill null values in the 'column_name' column with the previous value in the column
df['DIRECCION'].fillna(method='ffill', inplace=True)
# Fill null values in the 'column_name' column with the previous value in the column
df['DISTRITO'].fillna(method='ffill', inplace=True)
```

Utilizamos el método fillna para reemplazar los datos NaN del dataframe. El método que utilizamos para dicho reemplazo es el fill forward; este método sustituye el valor faltante por el valor de la última observación disponible.

```python
# Codificar variables categóricas utilizando one-hot encoding
df_encoded = pd.get_dummies(df, columns=['DEPARTAMENTO', 'MUNICIPIO', 'ESTABLECIMIENTO', 'DIRECCION', 'NIVEL', 'SECTOR', 'AREA', 'STATUS', 'MODALIDAD', 'JORNADA', 'PLAN', 'DEPARTAMENTAL'])
```

Utilizamos el método get_dummies para convertir las columnas de variables categóricas a columnas con variables dummies -con valores 1,0-.

```python
# vemos los valores unicos en cada columna del dataframe con el metodo .unique()
# guardamos en un diccionario. La llave es el nombre de la columna y la lista de valores unicos el el valor corresp
unique_values_dict = {}

# Loop through the columns and populate the dictionary
for column_name in df.columns:
    #print(len(df[column_name].unique()))
    unique_values_dict[column_name] = df[column_name].unique()
    #print("\n")

#print(unique_values_dict)

# lista de listas con los valores que queremos codificar
#list_of_lists = [unique_values_dict[column_name] for column_name in df.columns[df.columns.get_loc("SECTOR"):]]
list_of_lists = [unique_values_dict[column_name] for column_name in df.columns[df.columns.get_loc("SECTOR"):df.columns.get_loc(df.columns[-1])]]

#print(df.columns.get_loc("SECTOR"))

#list_of_lists
```

```python
# vemos cuantos valores unicos existen por columna del dataframe
for key in unique_values_dict:
    print(key + " " + str(len(unique_values_dict[key])))
```

Al observar los valores unicos por columna del dataframe, vemos que todas las escuelas tienen nivel diversificado, por lo tanto podemos eliminar dicha columna del dataframe. Por lo otro lado, vamos a codificar el sector, area, status, modalidad, jornada y plan.

La manera de codificar es la siguiente; cada columna tiene una lista de valores únicos y vamos a colocar como código la posición del dato observado, la respectiva posicion que ocupa dicho dato en la lista de valores unicos.

```python
# hacemos la codificacion con el area
unique_values_area = unique_values_dict["SECTOR"]
# Create a mapping of values to positions in the unique values list
value_to_position = {value: index for index, value in enumerate(unique_values_area)}

df['encoded_SECTOR'] = df['SECTOR'].map(value_to_position)

# eliminamos la columna area despues de codificarla
df.drop('SECTOR', axis=1, inplace=True)

#df
```

```python
# hacemos la codificacion con el area
unique_values_area = unique_values_dict["AREA"]
# Create a mapping of values to positions in the unique values list
value_to_position = {value: index for index, value in enumerate(unique_values_area)}

df['encoded_AREA'] = df['AREA'].map(value_to_position)

# eliminamos la columna area despues de codificarla
df.drop('AREA', axis=1, inplace=True)

#df
```

```python
# hacemos la codificacion con el status
unique_values_area = unique_values_dict["STATUS"]
# Create a mapping of values to positions in the unique values list
value_to_position = {value: index for index, value in enumerate(unique_values_area)}

df['encoded_STATUS'] = df['STATUS'].map(value_to_position)

# eliminamos la columna area despues de codificarla
df.drop('STATUS', axis=1, inplace=True)

#df
```

```python
# hacemos la codificacion con la modalidad
unique_values_area = unique_values_dict["MODALIDAD"]
# Create a mapping of values to positions in the unique values list
value_to_position = {value: index for index, value in enumerate(unique_values_area)}

df['encoded_MODALIDAD'] = df['MODALIDAD'].map(value_to_position)

# eliminamos la columna area despues de codificarla
df.drop('MODALIDAD', axis=1, inplace=True)

#df
```

```python
# hacemos la codificacion con la jornada
unique_values_area = unique_values_dict["JORNADA"]
# Create a mapping of values to positions in the unique values list
value_to_position = {value: index for index, value in enumerate(unique_values_area)}

df['encoded_JORNADA'] = df['JORNADA'].map(value_to_position)

# eliminamos la columna area despues de codificarla
df.drop('JORNADA', axis=1, inplace=True)
```

```python
# hacemos la codificacion con el PLAN
unique_values_area = unique_values_dict["PLAN"]
# Create a mapping of values to positions in the unique values list
value_to_position = {value: index for index, value in enumerate(unique_values_area)}

df['encoded_PLAN'] = df['PLAN'].map(value_to_position)

# eliminamos la columna area despues de codificarla
df.drop('PLAN', axis=1, inplace=True)

# puesto que todos los colegios tiene el nivel diversificado, eliminamos dicha columna
df.drop('NIVEL', axis=1, inplace=True)

# puesto que la columna departamento ya incluye esta informacion, eliminamos dicha columna
df.drop('DEPARTAMENTAL', axis=1, inplace=True)

# considerando que la direccion del colegio no es necesario para el analisis, eliminamos dicha columna
df.drop('DIRECCION', axis=1, inplace=True)

df
```
